## Cosine Similarity

```python
preference_vector = np.zeros(46) 
for prefs in user_preferences.values():
    for feature in prefs:
        preference_index = int(feature)
        preference_vector[preference_index] += 1

original_indices = [1, 4, 5, 6, 7, 8, 12, 16, 20, 21, 25, 26, 29, 40, 43]        
preference_vector = preference_vector[original_indices]
# Normalize preference vector to weight preferences proportionally
preference_vector /= max(preference_vector)
valid_businesses = set()
for user_id in user_preferences.keys():
    if user_id in top_rated['user_id'].values:
        user_top_businesses = top_rated.loc[top_rated['user_id'] == user_id, 'top_businesses'].values[0]
    else:
        print(f"No data available for user_id {user_id}")
        user_top_businesses = top_rated[top_rated['user_id'] == 'new_user']
    user_top_businesses = set(user_top_businesses)
    if not valid_businesses:
        valid_businesses = user_top_businesses
    else:
        valid_businesses = valid_businesses.intersection(user_top_businesses)

filtered_restaurants = feature_df[feature_df['business_id'].isin(valid_businesses)]

# Construct a feature matrix from these restaurants
feature_columns = [f'{i}' for i in original_indices]  # Adjust if column names are different
restaurant_features = filtered_restaurants[feature_columns].to_numpy()
# Compute cosine similarity between aggregated user preference vector and each restaurant's features
similarities = cosine_similarity([preference_vector], restaurant_features)[0]

# Get the indices of the top 5 matching restaurants
top_5_indices = np.argsort(similarities)[-5:][::-1]

# Output the top 5 matches
print("\nTop 5 matches based on aggregated user preferences: \n")
for index in top_5_indices:
    best_match = filtered_restaurants.iloc[index]
    similarity_score = similarities[index]
    print("Business ID: ", best_match['business_id'])
    print("Name: ", best_match['name'])
    print("Address: ", best_match['address'], best_match.get('postal_code', ''))
    print("Similarity Score: {:.4f}".format(similarity_score))
    print()
```

## Binary BM25

```python
def binary_bm25(restaurant_features, user_prefs_vector, k1=1.5, b=0.75):
    # Calculate document lengths for the BM25 length normalization
    doc_lengths = np.sum(restaurant_features, axis=1)
    avg_doc_length = np.mean(doc_lengths)
    
    # Calculate IDF for each feature in user preferences
    N = restaurant_features.shape[0]
    df = np.sum(restaurant_features, axis=0)
    idf = np.log((N - df + 0.5) / (df + 0.5) + 1)  # Added 1 to prevent negative values
    
    # Apply the binary BM25 formula
    tf = restaurant_features * user_prefs_vector  # binary TF (either 0 or 1)
    # Adjust tf calculation for proper broadcasting
    tf = (tf * (k1 + 1)) / (tf + k1 * (1 - b + b * (doc_lengths[:, np.newaxis] / avg_doc_length)))
    bm25_scores = np.sum(tf * idf, axis=1)  # Element-wise multiplication with idf
    
    return bm25_scores

def predict_restaurant(user_preferences, top_rated, feature_df):
    # Assuming binary features are already calculated
    # preference_vector and other code setup remains as previously defined
    
    user_prefs_vector = np.zeros(46) 
    for prefs in user_preferences.values():
        for feature in prefs:
            preference_index = int(feature)
            user_prefs_vector[preference_index] += 1

    original_indices = [1, 4, 5, 6, 7, 8, 12, 16, 20, 21, 25, 26, 29, 40, 43]        
    user_prefs_vector = user_prefs_vector[original_indices]

    valid_businesses = set()
    for user_id in user_preferences.keys():
        if user_id in top_rated['user_id'].values:
            user_top_businesses = top_rated.loc[top_rated['user_id'] == user_id, 'top_businesses'].values[0]
        else:
            print(f"No data available for user_id {user_id}")
            user_top_businesses = top_rated[top_rated['user_id'] == 'new_user']
        user_top_businesses = set(user_top_businesses)
        if not valid_businesses:
            valid_businesses = user_top_businesses
        else:
            valid_businesses = valid_businesses.intersection(user_top_businesses)
    
    filtered_restaurants = feature_df[feature_df['business_id'].isin(valid_businesses)]
    # Construct a feature matrix from these restaurants
    feature_columns = [f'{i}' for i in original_indices]  # Adjust if column names are different
    restaurant_features = filtered_restaurants[feature_columns].to_numpy()

    # Compute BM25 scores
    bm25_scores = binary_bm25(restaurant_features, user_prefs_vector)
    
    # Get the indices of the top 5 matching restaurants
    top_5_indices = np.argsort(bm25_scores)[-5:][::-1]

    # Output the top 5 matches
    print("\nTop 5 matches based on aggregated user preferences: \n")
    for index in top_5_indices:
        best_match = filtered_restaurants.iloc[index]
        similarity_score = bm25_scores[index]
        print("Business ID: ", best_match['business_id'])
        print("Name: ", best_match['name'])
        print("Address: ", best_match['address'], best_match.get('postal_code', ''))
        print("Similarity Score: {:.4f}".format(similarity_score))
        print()
```

## Jaccard Similarity

```python
def jaccard_similarity(vector1, matrix):
    # Ensure binary representation (0s and 1s only)
    vector1 = np.clip(vector1, 0, 1)
    matrix = np.clip(matrix, 0, 1)
    
    intersection = np.logical_and(matrix, vector1).sum(axis=1)
    union = np.logical_or(matrix, vector1).sum(axis=1)
    similarity = intersection / union
    return similarity

def predict_restaurant(user_preferences, top_rated, feature_df):
    
    preference_vector = np.zeros(46) 
    for prefs in user_preferences.values():
        for feature in prefs:
            preference_index = int(feature)
            preference_vector[preference_index] += 1

    original_indices = [1, 4, 5, 6, 7, 8, 12, 16, 20, 21, 25, 26, 29, 40, 43]        
    preference_vector = preference_vector[original_indices]
    # Normalize preference vector to weight preferences proportionally
    preference_vector /= max(preference_vector)
    valid_businesses = set()
    for user_id in user_preferences.keys():
        if user_id in top_rated['user_id'].values:
            user_top_businesses = top_rated.loc[top_rated['user_id'] == user_id, 'top_businesses'].values[0]
        else:
            print(f"No data available for user_id {user_id}")
            user_top_businesses = top_rated[top_rated['user_id'] == 'new_user']
        user_top_businesses = set(user_top_businesses)
        if not valid_businesses:
            valid_businesses = user_top_businesses
        else:
            valid_businesses = valid_businesses.intersection(user_top_businesses)
    
    filtered_restaurants = feature_df[feature_df['business_id'].isin(valid_businesses)]

    # Construct a feature matrix from these restaurants
    feature_columns = [f'{i}' for i in original_indices]  # Adjust if column names are different
    restaurant_features = filtered_restaurants[feature_columns].to_numpy() 
    
    similarities = jaccard_similarity(preference_vector, restaurant_features)
    
    # similarities = cosine_similarity([preference_vector], restaurant_features)[0]

    # Get the indices of the top 5 matching restaurants
    top_5_indices = np.argsort(similarities)[-5:][::-1]

    # Output the top 5 matches
    print("\nTop 5 matches based on aggregated user preferences: \n")
    for index in top_5_indices:
        best_match = filtered_restaurants.iloc[index]
        similarity_score = similarities[index]
        print("Business ID: ", best_match['business_id'])
        print("Name: ", best_match['name'])
        print("Address: ", best_match['address'], best_match.get('postal_code', ''))
        print("Similarity Score: {:.4f}".format(similarity_score))
        print()
```
