```python
import pandas as pd

restaurants = pd.read_csv('restaurants.csv')
restaurants.head()
```

```python
restaurants_phili = restaurants[restaurants['city'] == 'Philadelphia']
restaurants_phili = restaurants[['name','business_id','categories','address','stars']]
print(restaurants_phili)
```

```python
restaurants_phili['categories'] = restaurants_phili['categories'].apply(lambda x: [s.strip() for s in x.split(',')])
```

```python
print(restaurants_phili.iloc[0]['categories'])
```

```python
# new_features file contains mapping of 256 features to a smaller subset based on groupings

feature_map = pd.read_csv('new_features.csv')
feature_map['Feature'] = feature_map['Feature'].str.replace("',", "")
feature_map.head()
```

```python
feature_map.head()

feature_map = feature_map[feature_map['value1'] != 'drop']
feature_map = feature_map[feature_map['value1'] != 'drop?']
feature_map = feature_map.dropna(subset=['value1'])


feature_map['value1'] = feature_map['value1'].astype(int)
```

```python
# we grouped some more features together so next few blocks of code are grouping 28 and 17 together, 6 and 25, and 14 and 7
for i in range(len(feature_map)):
    if feature_map.iloc[i]['value1'] == 28:
        feature_map.iloc[i, feature_map.columns.get_loc('value1')] = 17
```

```python

for i in range(len(feature_map)):
    if feature_map.iloc[i]['value1'] == 25:
        feature_map.iloc[i, feature_map.columns.get_loc('value1')] = 6
```

```python
for i in range(len(feature_map)):
    if feature_map.iloc[i]['value1'] == 14:
        feature_map.iloc[i, feature_map.columns.get_loc('value1')] = 7
```

```python
for i in range(len(feature_map)):
    if feature_map.iloc[i]['value1'] > 28:
        feature_map.iloc[i, feature_map.columns.get_loc('value1')] = feature_map.iloc[i, feature_map.columns.get_loc('value1')] - 1
```

```python
for i in range(len(feature_map)):
    if feature_map.iloc[i]['value1'] > 25:
        feature_map.iloc[i, feature_map.columns.get_loc('value1')] = feature_map.iloc[i, feature_map.columns.get_loc('value1')] - 1
```

```python
for i in range(len(feature_map)):
    if feature_map.iloc[i]['value1'] > 14:
        feature_map.iloc[i, feature_map.columns.get_loc('value1')] = feature_map.iloc[i, feature_map.columns.get_loc('value1')] - 1
```

```python
for i in range(len(feature_map)):
    if feature_map.iloc[i]['value2'] > 28:
        feature_map.iloc[i, feature_map.columns.get_loc('value2')] = feature_map.iloc[i, feature_map.columns.get_loc('value2')] - 3
```

```python
feature_mapping = {}
for i in range(len(feature_map)):
  vals = []
  if not pd.isna(feature_map.iloc[i]['value1']) and feature_map.iloc[i]['value1']!="drop" and feature_map.iloc[i]['value1']!="drop?":
    vals.append(feature_map.iloc[i]['value1'])
  if not pd.isna(feature_map.iloc[i]['value2']):
    vals.append(feature_map.iloc[i]['value2'])
  if not pd.isna (feature_map.iloc[i]['value3']):
    vals.append(feature_map.iloc[i]['value3'])

  feature_mapping[feature_map.iloc[i]['Feature']] = vals
```

```python
print(feature_mapping)
```

```python
cols = ['ID','Restaurant']
for i in range (1,47):
  cols.append(i)

updated_restaurants = pd.DataFrame(columns=cols)

for i in range(len(restaurants_phili)):
  row = restaurants_phili.iloc[i]
  categs = row['categories']
  counts = {}
  new_row = []
  new_row.append(row['business_id'])
  restaurant_name = row['name']
  for value in categs:
    if not pd.isna(value):
      if value in feature_mapping:
        new_val = feature_mapping[value]
        #print(new_val)
        for v in new_val:
          counts[int(v)]=1
  new_row.append(restaurant_name)
  for j in range(1,47):
    if j in counts:
      new_row.append(counts[j])
    else:
      new_row.append(0)
  updated_restaurants.loc[i]=new_row
```

```python
updated_restaurants.head()
updated_restaurants.to_csv('restaurants_new.csv', index=False)

```
